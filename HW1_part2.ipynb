```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

```python
image_location = "/Users/tianyuwang/Documents/UCSD/Spring2017/CSE152/HW1/data/ucsd_logo.png"
img = cv2.imread(image_location,1)
img.shape
```

```python
# helper function to rotate the image clockwise by 90 degrees (pi/2 rad)
def rotate_right(image):
    new_image = np.zeros((1000, 1000, 3))
    height = np.size(img, 0)
    width = np.size(img, 1)
    for i in range(height):
        for j in range(width):
            new_image[j,(width-i-1),:] = image[i,j,:]
    return new_image
```

```python
b,g,r = cv2.split(img)
img1 = cv2.merge([r,g,b])
img2 = rotate_right(img1)
img3 = rotate_right(img2)
img4 = rotate_right(img3)
```

```python
img2.shape
```

```python
f, axarr = plt.subplots(2, 2)
axarr[0, 0].imshow(img1)
axarr[0, 1].imshow(img2)
axarr[1, 0].imshow(img3)
axarr[1, 1].imshow(img4)
plt.setp([a.get_xticklabels() for a in axarr[0, :]], visible=False)
plt.setp([a.get_yticklabels() for a in axarr[:, 1]], visible=False)
plt.show()
```
